#### Apakah ada tidaknya describsi mempengaruhi viewer pada video yang di publish di Negara US ?

Disini kita akan mengujian hipotesis ada tidaknya describsi mempengaruhi viewer pada video yang di publish di Negara US dengan Two Sample Independent T-Test dikarenakan data tidak terdistribusi normal. Untuk pembuktian data tidak terdistribusi normal sudah dilakukan dua cara diatas yaitu Grapical Methods (Histogram) dan Frequesntist Test (D'Agostino and Pearson's Test)

```python
df_1[['views', 'Description_1']].head()
```

```python
# Statistik Test
# Apakah ada tidaknya describsi mempengaruhi viewer pada video yang di publish di Negara US ?
# Dikarenakan data tidak terdistribusi secara normal, kita akan menggunakan Two Sample Independent T-Test
#Import Library
from scipy.stats import ttest_ind

#Two sample independent t-test
result = list(ttest_ind(
              df_1[df_1['Description_1'] == 0]['views'],
              df_1[df_1['Description_1'] == 1]['views'],
              equal_var = False))

print('T-Statistic:', result[0])
print('P-Value:', result[1])

#Conclusion
if result[1]/2 < 0.05:
    print('Berhasil Menolak H0,''\n kita punya cukup bukti bahwa rata rata views tanpa deskirpsi lebih sedikit dari pada dengan deskripsi')
else:
    print('Gagal Menolak H0,' '\n kita tidak punya cukup bukti bahwa rata rata views tanpa deskirpsi lebih sedikit dari pada dengan deskripsi')
```

```python
# Statistik Test
# Apakah ada tidaknya describsi mempengaruhi viewer pada video yang di publish di Negara US ?
# Dikarenakan data tidak terdistribusi secara normal, kita akan menggunakan Two Sample Independent T-Test
#Import Library
from scipy.stats import ttest_ind

#Two sample independent t-test
result = list(ttest_ind(
              df_1[df_1['tag_1'] == 0]['views'],
              df_1[df_1['tag_1'] == 1]['views'],
              equal_var = False))

print('T-Statistic:', result[0])
print('P-Value:', result[1])

#Conclusion
if result[1]/2 < 0.05:
    print('Berhasil Menolak H0,''\n kita punya cukup bukti bahwa rata rata views tanpa tags lebih sedikit dari pada dengan tags')
else:
    print('Gagal Menolak H0,' '\n kita tidak punya cukup bukti bahwa rata rata views tanpa tags lebih sedikit dari pada dengan tags')
```

```python
# Import Library
from scipy.stats import mannwhitneyu

# Mann WHitney Test
result = list(mannwhitneyu(
    df_1[df_1['label_trending'] == 0]['Day'],
    df_1[df_1['label_trending'] == 1]['Day'],
    alternative = 'greater' # M (no read) > M (read)
))

print('U-Stactistic', result[0])
print('P-Value', result[1])

if result[1] < 0.05:
    print('Berhasil Menolak H0,''\n Median video view yang trending yang butuh waktu dalam hari lama  <  Median video view yang trending yang butuh waktudalam hari cepat')
else:
    print('Gagal Menolak H0,' '\n Median video view yang trending yang butuh waktu dalam hari lama =  Median video view yang trending yang butuh waktu dalam hari cepat')
```
